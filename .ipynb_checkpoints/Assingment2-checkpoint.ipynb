Computation of gradients

```python
# Compute the gradients of the cost function for a mini-batch    
def ComputeGradients(X, y, Y, W1, W2, B1, B2, lamda, mode):
    grad_W1=np.zeros(W1.shape)
    grad_W2=np.zeros(W2.shape)
    grad_b1=np.zeros(B1.shape)
    grad_b2=np.zeros(B2.shape)
    batchSize=len(y)
    
    if DropOut is 'on':
            X_DropOutMask=np.random.binomial(1,1-DropOutRate/100,X.shape)
            X *= X_DropOutMask
    
    for i in range(batchSize):
        w1 = np.copy(W1)
        w2 = np.copy(W2)
        b1 = np.copy(B1)
        b2 = np.copy(B2)
        Xi=np.expand_dims(X[:,i], axis=1)
        Yi=np.expand_dims(Y[:,i], axis=1)         
        
        
        if DropConnect is 'on':
            W1_DropConnectMask=np.random.binomial(1,1-DropOutRate/100,W1.shape)
            W2_DropConnectMask=np.random.binomial(1,1-DropOutRate/100,W2.shape)
            w1 *= W1_DropConnectMask
            w2 *= W2_DropConnectMask

        if mode is 'ReLu':
            # pass 1st layer
            s1 = fwd(Xi,w1,b1)
            # pass activation
            h = ActFunction(s1, mode)
            if DropOut is 'on':
                h_DropOutMask=np.random.binomial(1,1-DropOutRate/100,h.shape)
                h *= h_DropOutMask
            #pass 2nd layer
            s = fwd(h,w2,b2)
            #pass activation
            p = ActFunction(s, 'SoftMax')
            # Compute g
            g=-np.dot(Yi.T/Yi.T.dot(p),np.diag(p.flatten())-p.dot(p.T))
            # Add gradient of loss w.r.t. second layer parameters computed at (x, y)
            grad_b2 += g.T
            grad_W2 += np.dot(g.T,h.T)
        
            # Propagate the gradients
            g=np.dot(g,w2)
            ind=1*(s1>0).flatten()
            g=np.dot(g,np.diag(ind))
            # Add gradient of l w.r.t. first layer parameters computed at (x; y)
            grad_b1 += g.T
            grad_W1 += np.dot(g.T,Xi.T)
        
        elif mode is 'LeakyReLu':
            # pass 1st layer
            s1 = fwd(Xi,w1,b1)
            # pass activation
            h = ActFunction(s1, mode)
            if DropOut is 'on':
                h_DropOutMask=np.random.binomial(1,1-DropOutRate/100,h.shape)
                h *= h_DropOutMask
            #pass 2nd layer
            s = fwd(h,w2,b2)
            #pass activation
            p = ActFunction(s, 'SoftMax')
            # Compute g
            g=-np.dot(Yi.T/Yi.T.dot(p),np.diag(p.flatten())-p.dot(p.T))
            # Add gradient of loss w.r.t. second layer parameters computed at (x, y)
            grad_b2 += g.T
            grad_W2 += np.dot(g.T,h.T)
            
            # Propagate the gradients
            g=np.dot(g,w2)
            ind=((s1>0)*1+(s1<0)*0.01).flatten()
            g=np.dot(g,np.diag(ind))
            # Add gradient of l w.r.t. first layer parameters computed at (x; y)
            grad_b1 += g.T
            grad_W1 += np.dot(g.T,Xi.T)
        
    #Divide by the number of entries in batch to get average
    grad_W1 /= batchSize
    grad_W2 /= batchSize
    grad_b1 /= batchSize
    grad_b2 /= batchSize
    
    # Add the gradient for the regularization term
    grad_W1 += 2*lamda*W1
    grad_W2 += 2*lamda*W2
    
    return grad_W1, grad_W2, grad_b1, grad_b2
```

The Mini-Batch algo

```python
# the minibatch algithm
def MiniBatchGD(X, y, Y, W1, W2, b1, b2, Vx, Vy, VY, GDparams, mode):
    
    # Create the v matrices
    vW1 = np.zeros(W1.shape)
    vW2 = np.zeros(W2.shape)
    vb1 = np.zeros(b1.shape)
    vb2 = np.zeros(b2.shape)
    
    # define statistic lists
    lossOnTrain=[]
    lossOnValidation=[]
    costOnTrain=[]
    costOnValidation=[]
    accuracyOnTrain=[]
    accuracyOnValidation=[]
    storedW1=[]
    storedW2=[]
    storedb1=[]
    storedb2=[]
    
    #Compute initial stats for thaining and validation set
    loss, cost = ComputeCost(X, y, Y, W1, W2, b1, b2, GDparams.lamda, mode)
    lossOnTrain.append(loss)
    costOnTrain.append(cost)
    loss, cost = ComputeCost(Vx, Vy, VY, W1, W2, b1, b2, GDparams.lamda, mode)
    lossOnValidation.append(loss)
    costOnValidation.append(cost)
    accuracyOnTrain.append(ComputeAccuracy(X, y, W1, W2, b1, b2, mode))
    accuracyOnValidation.append(ComputeAccuracy(Vx, Vy, W1, W2, b1, b2, mode))

    #start mini-batch
    for epoch in range(GDparams.n_epochs):
        i=0 

        if RandOrder is 'on' and epoch > 1:
            reOrder = np.random.permutation(len(y))
            X=X[:,reOrder] 
            Y=Y[:,reOrder]
            y=y[reOrder] 
        
        while i<len(X.T):
            ### create batch
            if i+GDparams.n_batch > len(y):
                Xbatch=X[:,i:len(y)]
            else:
                Xbatch=X[:,i:i+GDparams.n_batch]
            if i+GDparams.n_batch > len(y):
                Ybatch=Y[:,i:len(y)]
            else:
                Ybatch=Y[:,i:i+GDparams.n_batch] 
            if i+GDparams.n_batch > len(y):
                ybatch=y[i:len(y)]
            else:
                ybatch=y[i:i+GDparams.n_batch] 
            
            if Jitter is 'on' and epoch > 1:
                Xbatch += np.random.normal(0, 0.01 * np.std(Xbatch) ,Xbatch.shape)  
                
            #-----> TRAIN BATCH ->            
            if mode is 'ReLu': 
                grad_W1, grad_W2, grad_b1, grad_b2 = ComputeGradients(Xbatch, ybatch, Ybatch, W1, W2, b1, b2,\
                                                                      GDparams.lamda, mode)
                
            if mode is 'LeakyReLu': 
                grad_W1, grad_W2, grad_b1, grad_b2 = ComputeGradients(Xbatch, ybatch, Ybatch, W1, W2, b1, b2,\
                                                                      GDparams.lamda, mode)   
                
            elif mode is 'AnCal':
                grad_W1, grad_W2, grad_b1, grad_b2 = ComputeGradientsNum(Xbatch, ybatch, Ybatch, W1, W2, b1, b2,\
                                                                         GDparams.lamda)
            
            # update the weights
            W1, W2, b1, b2, vW1, vW2, vb1, vb2 = WeightUpdate(Xbatch, ybatch, Ybatch, W1, W2, b1, b2,\
                                                              grad_W1, grad_W2, grad_b1, grad_b2, \
                                                                      GDparams, vW1, vW2, vb1, vb2, mode)

            #move to the next batch
            i+=GDparams.n_batch

        #set drop learing rate to on or off
        if DropRate=='on':
            GDparams.eta-=GDparams.eta-GDparams.eta*0.9
            
        #Compute stats for thaining and validation set durring learning
        
        loss, cost = ComputeCost(X, y, Y, W1, W2, b1, b2, GDparams.lamda, mode)
        lossOnTrain.append(loss)
        costOnTrain.append(cost)
        loss, cost = ComputeCost(Vx, Vy, VY, W1, W2, b1, b2, GDparams.lamda, mode)
        lossOnValidation.append(loss)
        costOnValidation.append(cost)
        accuracyOnTrain.append(ComputeAccuracy(X, y, W1, W2, b1, b2, mode))
        accuracyOnValidation.append(ComputeAccuracy(Vx, Vy, W1, W2, b1, b2, mode))
        
        # store the weights
        storedW1.append(W1)
        storedW2.append(W2)
        storedb1.append(b1)
        storedb2.append(b2)
        
        #set early stopping
        if EarlyStop is 'on' and epoch > 10 :
            TrackAcc=np.average(accuracyOnValidation[epoch-10:epoch])
            if TrackAcc > accuracyOnValidation[epoch+1]:
                break
        
        #set breaking point for the training/validating cost exploding
        if costOnTrain[epoch+1] > 3 * costOnTrain[0] or costOnValidation[epoch+1] > 3 *costOnValidation[0]:
            break
    BestIter=np.argmax(accuracyOnValidation)
    
    return storedW1[BestIter-1], storedW2[BestIter-1], storedb1[BestIter-1], storedb2[BestIter-1],\
            lossOnTrain, lossOnValidation, costOnTrain, costOnValidation, accuracyOnTrain, accuracyOnValidation, BestIter
```

weight update

```python
# different wheight update manners
def WeightUpdate(X, y, Y, W1, W2, b1, b2, grad_W1, grad_W2, grad_b1, grad_b2, GDparams, vW1, vW2, vb1, vb2, mode):
    
    gamma=0.9

    if UpdateMode is 'Simple':
        # Simple subtraction of the gradient
        
        W1 -= GDparams.eta * grad_W1
        W2 -= GDparams.eta * grad_W2
        b1 -= GDparams.eta * grad_b1
        b2 -= GDparams.eta * grad_b2
        
    elif UpdateMode is 'Momentum':
        # SGD with momentum
        
        vW1 = gamma * vW1 + GDparams.eta * grad_W1
        vW2 = gamma * vW2 + GDparams.eta * grad_W2
        vb1 = gamma * vb1 + GDparams.eta * grad_b1
        vb2 = gamma * vb2 + GDparams.eta * grad_b2
        
        W1 -= vW1
        W2 -= vW2
        b1 -= vb1
        b2 -= vb2

    elif UpdateMode is 'NAG':
        # Nesterov accelerated gradient

        We1 = W1 - gamma * vW1
        We2 = W2 - gamma * vW2
        be1 = b1 - gamma * vb1
        be2 = b2 - gamma * vb2

        grad_We1, grad_We2, grad_be1, grad_be2 = ComputeGradients(X, y, Y, We1, We2, be1, be2, GDparams.lamda, mode)

        vW1 = gamma* vW1 + GDparams.eta * grad_We1
        vW2 = gamma* vW2 + GDparams.eta * grad_We2
        vb1 = gamma* vb1 + GDparams.eta * grad_be1
        vb2 = gamma* vb2 + GDparams.eta * grad_be2

        W1 -= vW1
        W2 -= vW2
        b1 -= vb1
        b2 -= vb2
    
    return W1, W2, b1, b2, vW1, vW2, vb1, vb2
```

Several functions for the algorithm

```python
# the object for the parameters of the minibatch
class GDparams(object):
    def __init__(self, lamda, n_epochs, n_batch, eta, nodes):
        self.lamda=lamda
        self.n_epochs=n_epochs
        self.n_batch=n_batch
        self.eta=eta
        self.nodes=nodes

# READ THE DATA function
def LoadBatch(filename):
    #read the data from a CIFAR-10 batch file and return the image and label data in separate files
    #this function is adopted from the source: https://github.com/PaddlePaddle/Paddle/issues/1301

    f = open(filename, 'rb')
    d = cPickle.load(f, encoding='bytes')
    d_decoded = {}
    for k, v in d.items():
        d_decoded[k.decode('utf8')] = v
    d = d_decoded
    f.close()
    data = d['data']
    labels = d['labels']
    filenames = [b.decode("utf-8") for b in d['filenames']]

    return data, labels, filenames


# preprocessing of the data to have 0 mean 
# and subtraction of this mean form the test and validation set
def PreProcessing(train, val, test):
    inSize=len(train.T)
    if inSize is not 1:
        TrainSetMean=np.expand_dims(np.mean(train,axis=1),1)
        train -=TrainSetMean
        val -=TrainSetMean
        test -=TrainSetMean
        
        return train, val, test
    else:
        return train, val, test


# Convert to One Hot Representation
def OneHotRep(labels):
    oneHot=np.zeros((len(labels),np.max(labels)+1))
    oneHot[np.arange(len(labels)), labels] = 1
    
    return oneHot

# weight initialization
def weightInt(inNodes,outNodes, mode,PCnumber):
    if mode is 'random':
        W=np.random.normal(0,0.001,(outNodes,inNodes))
#         b=np.random.normal(0,0.01,(outNodes,1))
        b=np.zeros((outNodes,1))
    
    if mode is 'PCA':
        W=np.zeros((outNodes,inNodes))
        b=np.random.normal(0,0.01,(outNodes,1))
        for i in range(trainingSetLabels.max()+1):
            #find all data from the wanted class
            pos=np.where(trainingSetLabels == i)
            #aply PCA on the dataset and export the 1st component
            c=trainingSet.T[pos].T
            pca = PCA(PCnumber, 'full')
            DimRed = pca.fit_transform(c).T
            for comp in range(PCnumber):
                W[i]+=DimRed[comp]
        W=((W)/(np.abs(W.min())+W.max()))/100.
    return W,b

# visualize the learning stats
def LearningStats(params,lossOnTrain,lossOnValidation,costOnTrain,costOnValidation,accuracyOnTrain,accuracyOnValidation,name):
    color1='r'
    color2='g'
    fig = plt.figure()
    #plot the loss
    ax = fig.add_subplot(1, 3, 1)
    ax.plot(lossOnTrain, color1, label='Training Set')
    ax.plot(lossOnValidation,color2, label='Validation Set')
    ax.set_title('loss vs iter')
    ax.set_xlabel('Epoch')
    ax.set_ylabel('loss')
    
     #plot the cost
    ax = fig.add_subplot(1, 3, 2)
    ax.plot(costOnTrain,color1, label='Training Set')
    ax.plot(costOnValidation,color2, label='Validation Set')
    ax.set_title('cost vs iter')
    ax.set_xlabel('Epoch')
    ax.set_ylabel('cost')
    
    #plot accuracy
    ax = fig.add_subplot(1, 3, 3)
    ax.plot(accuracyOnTrain,color1, label='Training Set')
    ax.plot(accuracyOnValidation,color2, label='Validation Set')
    ax.set_title('accuracy vs iter')
    ax.set_xlabel('Epoch')
    ax.set_ylabel('accuracy')
    plt.legend(bbox_to_anchor=(-1, 1), loc='upper left', ncol=1)
    
    plt.subplots_adjust(left=-1.2)
    # lamda, n_epochs, n_batch, eta 
    plt.savefig('results/LearningStats_lambda=%s_batch=%s_%s.png'\
                %(params.lamda,params.n_batch,name), bbox_inches='tight')
    plt.show()
    
# weight visualization
def VisWeights(w,params,name):
    grb=3
    images=[]
    for img in w:
        dim=np.sqrt(img.size/grb)
        rawImage=np.rot90(img.reshape(dim, dim, grb),-1)
        image=((rawImage-np.min(rawImage))/(np.max(rawImage)-np.min(rawImage)))
        images.append(image)

    fig = plt.figure(figsize=(20, 10))
    fig.suptitle('First Layer\'s Weight Visualization', fontsize=26, fontweight='bold')
    
    if len(w)%10 > 6:
        rowSize=round(len(w)/10)
    else:
        rowSize=round(len(w)/10)+1

    for idx in range(len(images)):
            ax = fig.add_subplot(rowSize, 10,idx+1)
            ax.imshow(images[idx])
            ax.axis('off')
    plt.savefig('results/WeightVisualization_lambda=%s_batch=%s_%s.png'\
                %(params.lamda,params.n_batch,name), bbox_inches='tight')
    plt.show()
    
# forward the input to output through matrix multiplication with weights and the addition of bias
def fwd(X,W,b):
    s=np.dot(W,X)+b
    
    return s

# the activation functions
def ActFunction(s, function):
    if function is 'SoftMax':
        probability=np.exp(s)/np.sum(np.exp(s), axis=0)
        return probability
    elif function is 'ReLu':
        out = np.maximum(0,s)
        return out
    elif function is 'LeakyReLu':
        out = np.maximum(0.01 * s,s)
        return out
    else:
        print("Error in activation function")
        
# the network evaluation i.e. network's output
def NetOut(X, W1, W2, b1, b2, mode, status='None'):
    # pass 1st layer
    s1 = fwd(X,W1,b1)
    # pass activation
    h = ActFunction(s1, mode)
    #pass 2nd layer
    s = fwd(h,W2,b2)
    #pass activation
    p = ActFunction(s, 'SoftMax')

    if status is 'leaky':
        return s1, h, s, p
    else:
        return p


# compute the cost function
def ComputeCost(X, y, Y, W1, W2, b1, b2, lamda, mode):
    NetEval=NetOut(X, W1, W2, b1,b2, mode)
    crossEntropy=np.diag(-np.log(np.dot(Y.T,NetEval)))
    regularization = lamda*(np.sum(W1**2)+np.sum(W2**2))
    loss=np.sum(crossEntropy)/len(y)
    cost=loss + regularization
    
    return loss, cost
 


# compute the net's accuracy
def ComputeAccuracy(X, y, W1, W2, b1, b2, mode):
    NetEval=NetOut(X, W1, W2, b1,b2, mode)
    k=np.argmax(NetEval, axis=0)
    diff=np.abs(k-y)
    accuracy=1-np.count_nonzero(diff)/len(y)

    return accuracy
```

The numerical computation of Gradients

```python

# Gradiants numerical computation
def ComputeGradsNumSlow(X, y, Y, W1, W2, b1, b2, lamda, mode, h=10**(-5)):

    # initialize gradients
    grad_W1=np.zeros(W1.shape)
    grad_b1=np.zeros(b1.shape)
    grad_W2=np.zeros(W2.shape)
    grad_b2=np.zeros(b2.shape)

    # compute gradients for bias (2nd layer)
    for i in range(len(b2)):
        b_try = np.copy(b2)
        b_try[i] -= h
        _,c1=ComputeCost(X, y, Y, W1, W2, b1, b_try, lamda, mode)
        b_try = np.copy(b2)
        b_try[i] += h
        b=b_try[i]
        _,c2=ComputeCost(X, y, Y, W1, W2, b1, b_try, lamda, mode)
        grad_b2[i] = (c2-c1) / (2*h)

    # compute gradients for bias (2nd layer)
    for i in range(len(W2)):
        for j in range(len(W2.T)):
            W_try = np.copy(W2)
            W_try[i,j] -= h
            _,c1=ComputeCost(X, y, Y, W1, W_try, b1, b2, lamda, mode)
            W_try = np.copy(W2)
            W_try[i,j] += h
            _,c2=ComputeCost(X, y, Y, W1, W_try, b1, b2, lamda, mode)
            grad_W2[i,j] = (c2-c1) / (2*h);
            
    # compute gradients for bias (1st layer)
    for i in range(len(b1)):
        b_try = np.copy(b1)
        b_try[i] -= h
        _,c1=ComputeCost(X, y, Y, W1, W2, b_try, b2, lamda, mode)
        b_try = np.copy(b1)
        b_try[i] += h
        b=b_try[i]
        _,c2=ComputeCost(X, y, Y, W1, W2, b_try, b2, lamda, mode)
        grad_b1[i] = (c2-c1) / (2*h)

    # compute gradients for bias (1st layer)
    for i in range(len(W1)):
        for j in range(len(W1.T)):
            W_try = W1.astype(float)
            W_try[i,j] -= h
            _,c1=ComputeCost(X, y, Y, W_try, W2, b1, b2, lamda, mode)
            W_try = W1.astype(float)
            W_try[i,j] += h
            _,c2=ComputeCost(X, y, Y, W_try, W2, b1, b2, lamda, mode)
            grad_W1[i,j] = (c2-c1) / (2*h);
    
    return grad_W1, grad_W2, grad_b1, grad_b2

# Compute the gradients of the cost function for a mini-batch    
def ComputeGradientsNum(X, y, Y, W1, W2, b1, b2, lamda, mode):
    grad_W1=np.zeros(W1.shape)
    grad_W2=np.zeros(W2.shape)
    grad_b1=np.zeros(b1.shape)
    grad_b2=np.zeros(b2.shape)
    batchSize=len(y)
    for i in range(batchSize):
        grad_W1num, grad_W2num, grad_b1num, grad_b2num =ComputeGradsNumSlow(X.astype(float),y, Y,\
                                    W1.astype(float), W2.astype(float), b1.astype(float), b2.astype(float), lamda, mode)
        grad_W1+=grad_W1num
        grad_W2+=grad_W2num
        grad_b1+=grad_b1num
        grad_b2+=grad_b2num
        
    #Divide by the number of entries in batch to get average
    grad_W1 /= batchSize
    grad_W2 /= batchSize
    grad_b1 /= batchSize
    grad_b2 /= batchSize
    
    return grad_W1, grad_W2, grad_b1, grad_b2

def ComputeGradDiff(X, y, Y, W1, W2, b1, b2, lamda, mode):
    #compute gradients numericaly
    grad_W1num, grad_W2num, grad_b1num, grad_b2num =ComputeGradientsNum(X, y, Y, W1, W2, b1, b2, lamda, mode)
    #compute gradients analyticaly
    grad_W1an, grad_W2an, grad_b1an, grad_b2an =ComputeGradients(X, y, Y, W1, W2, b1, b2, lamda, mode)
    #compute difference
    diffW1=np.abs(grad_W1an-grad_W1num)/np.max((np.abs(grad_W1an)+np.abs(grad_W1num)).clip(0))
    diffW2=np.abs(grad_W2an-grad_W2num)/np.max((np.abs(grad_W2an)+np.abs(grad_W2num)).clip(0))
    diffb1=np.abs(grad_b1an-grad_b1num)/np.max((np.abs(grad_b1an)+np.abs(grad_b1num)).clip(0))
    diffb2=np.abs(grad_b2an-grad_b2num)/np.max((np.abs(grad_b2an)+np.abs(grad_b2num)).clip(0))
    
    meanDiff=(diffW1.mean()+diffW2.mean()+diffb1.mean()+diffb2.mean())/2.
    
    print('The mean difference when computing the gradiants analyticaly and numericaly is %s'%meanDiff)
    print('in detail:')
    print('diffW1=%s, diffW2=%s, diffb1=%s, diffb2=%s'%(diffW1.mean(),diffW2.mean(),diffb1.mean(),diffb2.mean()))

# ComputeGradDiff(trainingSet, trainingSetLabels,trainingSetLabels_hot, wIn1, wIn2, bIn1, bIn2, 0, 'ReLu')
```
